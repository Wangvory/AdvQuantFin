# Diversification benefit in mean-variance portfolio optimization

```python
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np
```

```python
r_a = 0.04
r_b = 0.05
stdev_a = 0.07
stdev_b = 0.08

fig, ax = plt.subplots()
fig.set_size_inches(10, 8)

for correlation in np.linspace(-1, 1, 11):
    expected_returns, stdev_points = [], []
    for w_a in np.linspace(0, 1, 100):
        w_b = 1 - w_a
        r = r_a * w_a + r_b * w_b
        var = w_a**2 * stdev_a**2 + w_b**2 * stdev_b**2 + 2.0 * w_a * w_b * stdev_a * stdev_b * correlation
        stdev = np.sqrt(var)
        expected_returns.append(r)
        stdev_points.append(stdev)
    
    ax.plot(stdev_points, expected_returns, label=f'Correlation: {correlation:4.2f}')

ax.set_xlabel('Standard Deviation $\sigma$')
ax.set_ylabel('Expected return $r$')
ax.set_title('Diversification benefit for different levels of correlation between two assets')
ax.legend()

plt.show()
```
